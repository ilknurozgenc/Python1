```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image=cv2.imread("./deneme.jpg")    # ./ aynı dize içinde olduğunu gösterir
```

```python
type(image)
```

```python
image.shape  #görüntü piksellerini (en, boy, yükseklik) veriri.
```

```python
plt.imshow(image)   #opencv görüntüyü bgr formatında okur. yani renklerini değiştirir. bgr dan rgb ye dönüştürmek
```

```python
#1.kanal için bölme kesme işlemi

r,g,b=cv2.split(image)
print('r',r.shape)
print('g',g.shape)
print('b',b.shape)
```

```python
#yeniden şekillendirme, var olan boyutu küçültme

s=10
w=int(image.shape[1]*s/100)
h=int(image.shape[0]*s/100)
dim=(w,h)  #tupple tanımlama
re_size=cv2.resize(image,dim,interpolation=cv2.INTER_AREA)
re_size.shape
```

```python
#görüntüyü 90 derece döndürmek

(h,w)=image.shape[:2]

c=(w/2,h/2)

angle=90

m=cv2.getRotationMatrix2D(c,angle,1.0)
rotate_90=cv2.warpAffine(image,m,(h,w))
```

```python
plt.imshow(rotate_90)
```
